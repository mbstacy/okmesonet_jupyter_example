### OK Mesonet Example 

Use the github package to retrieve data from Mesonet.

##### Installation
   $ pip install https://github.com/mbstacy/okmesonet/zipball/master

```python
from okmesonet import weather
from datetime import datetime

start_date = datetime(2010,3,1)
end_date = datetime(2010,4,1)
site = "nrmn"

# Local Time
data =weather.get_mesonet_dataframe(start_date,end_date,site,local_time=True) 
```

#### View OK Mesonet Data 
This object is a pandas dataframe

```python
data
```

#### Quick Plot Example


```python
grp = data.groupby(['DATE'])
grp=grp.TAIR.agg(['min','mean','max'])
grp
```

```python
% matplotlib inline
grp=data.groupby(['DATE'])
grp= grp.agg(['min','mean','max'])
grp.TAIR.plot(rot=90)      
```
